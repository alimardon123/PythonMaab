```python
df = pd.DataFrame({'A': ['a', 'a', 'a', 'b', 'b', 'c', 'c'],
                  'B': [1, 5, 8, 4, 3, 2, 7],
                  'C': np.arange(7),
                  'E': ['v', 'v', 'v', 'c', 'c', 'c', 'c']})
df
```

```python
df.groupby('A')
```

```python
list(df.groupby('A'))
```

```python
df.groupby('A').sum(numeric_only=True)
```

```python
df.groupby(['E', 'A']).sum(numeric_only=True)
```

```python
df.groupby('A').agg({"C": 'sum', 'B':'mean'})
```

```python
df.groupby(['E', 'A']).agg('mean')
```

```python
df.groupby(['E', 'A']).agg(['mean', 'min'])
```

```python
df.groupby(['E', 'A']).agg({'B':['mean', 'max']})
```

```python
df.groupby(['E', 'A']).agg({'B':['mean', 'max']}).reset_index()
```

```python
df.groupby(['E', 'A']).agg(mean_B = pd.NamedAgg(column='B', aggfunc='mean'),
                          max_B = pd.NamedAgg(column='B', aggfunc='max'),
                          mean_C = pd.NamedAgg(column='C', aggfunc='mean'),
                          max_C = pd.NamedAgg(column='C', aggfunc='max'),).reset_index()
```

```python
list(df.groupby('A'))
```

```python
df.groupby('A').filter(lambda x : len(x)<3)
```

```python
df.groupby('A', as_index=False).apply(lambda x : x.nlargest(2, 'B').nsmallest(1, 'B'))
```
