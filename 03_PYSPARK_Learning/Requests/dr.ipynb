```python
import requests
payload = {"firstname": "Alex", "fname": 'Johnson'}
r  = requests.post('https://httpbin.org/post', payload)

print(r.text)
```

Cookies

```python
# In hhtpbin example

jokie = {'location' : 'New York'}

r = requests.post(url= 'https://httpbin.org/cookies', cookies=jokie)

print(r.text)
print(r.cookies.items())

```

```python
# Google Cookies

r2 = requests.get("https://google.com")
print(r2.cookies.get_dict())
```

```python
r = requests.get(url= 'https://httpbin.org/cookies', cookies=jokie)

print(r.text)
print(r.cookies.get_dict())
```

```python
url= 'https://dummyjson.com/products/'
num = 1
while True:
    
    if requests.get(url=url+str(num+1)).status_code ==404:
        r = requests.get(url=url+str(num))
        break
    num +=1
    print(num)
a = r.json()
print(a)
```

```python
r = requests.get(url= 'https://i.dummyjson.com/data/products/1/1.jpg')
with open('new.jpg', 'wb') as f:
    f.write(r.content)
```
